```python
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments
from datasets import load_dataset
raw_datasets = load_dataset("ashwinradhe/dl_dataset")
small_train_dataset = raw_datasets['train'].select(range(20000))
tokenizer = T5Tokenizer.from_pretrained("t5-small")
def tokenize_function(examples):
    model_inputs = tokenizer(examples["English Sentence"], max_length=512, padding="max_length", truncation=True)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples["French Sentence"], max_length=512, padding="max_length", truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
tokenized_datasets = small_train_dataset.map(tokenize_function, batched=True)
tokenized_datasets_1 = tokenized_datasets.train_test_split(test_size=0.2)
train_dataset = tokenized_datasets_1["train"]
eval_dataset = tokenized_datasets_1["test"]
test_dataset = eval_dataset.train_test_split(test_size=0.5)
valid_dataset = test_dataset["test"]
test_dataset = test_dataset["train"]
model = T5ForConditionalGeneration.from_pretrained("t5-small")
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=1000,
    save_strategy="steps"
)
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer,
)
trainer.train()
```

```python
import torch
def translate_sentences(model, tokenizer, sentences, device='cuda'):
    model.to(device)
    model.eval()
    translations = []
    for sentence in sentences:
        input_ids = tokenizer.encode(sentence, return_tensors='pt', max_length=512, truncation=True).to(device)
        with torch.no_grad():
            outputs = model.generate(input_ids, max_length=512, num_beams=5, early_stopping=True)
        translation = tokenizer.decode(outputs[0], skip_special_tokens=True)
        translations.append(translation)
    return translations
test_sentences = [
    "Hello, how are you?",
    "What is your name?",
    "I love learning new languages."
]
translations = translate_sentences(model, tokenizer, test_sentences)
for en, fr in zip(test_sentences, translations):
    print(f'English: {en}\nTranslated: {fr}\n')
```

```python
from datasets import load_metric
import torch

def calculate_bleu_score_batched(model, tokenizer, dataset, device="cuda", batch_size=4):
    model.to(device)
    model.eval()
    bleu_metric = load_metric("sacrebleu", trust_remote_code=True)
    predictions, references = [], []
    for i in range(0, len(dataset), batch_size):
        batch = dataset[i:i + batch_size]
        input_texts = tokenizer.batch_decode(batch['input_ids'], skip_special_tokens=True)
        labels = batch['labels']
        input_ids = tokenizer(input_texts, return_tensors="pt", padding=True, truncation=True, max_length=512).input_ids
        input_ids = input_ids.to(device)
        output_ids = model.generate(input_ids, max_length=512, num_beams=5, early_stopping=True)
        pred_texts = tokenizer.batch_decode(output_ids, skip_special_tokens=True)
        label_texts = tokenizer.batch_decode(labels, skip_special_tokens=True)
        for pred, label in zip(pred_texts, label_texts):
            bleu_metric.add(prediction=pred, reference=[label])
    final_score = bleu_metric.compute()
    return final_score["score"]
device = "cuda" if torch.cuda.is_available() else "cpu"
bleu_score = calculate_bleu_score_batched(model, tokenizer, test_dataset, device=device, batch_size=4)
print("BLEU Score:", bleu_score)
```

```python
from datasets import load_dataset

# Load the dataset
dataset = load_dataset("ashwinradhe/dl_dataset")

# Select the first 20000 rows
subset = dataset['train'].select(range(20000))

# Split the subset into train and test with a 80-20 ratio
train_test_split = subset.train_test_split(test_size=0.2)

# # The train_test_split now contains two datasets, one for training and one for testing
train_dataset = train_test_split['train']
testing_dataset = train_test_split['test']
test_val_split = testing_dataset.train_test_split(test_size=0.5)
test_dataset = test_val_split['train']
val_dataset = test_val_split['test']
```

##Facebook M2m

```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
from transformers import DataCollatorForSeq2Seq, Seq2SeqTrainingArguments, Seq2SeqTrainer
model_name = "facebook/m2m100_418M"
tokenizer = M2M100Tokenizer.from_pretrained(model_name)
model = M2M100ForConditionalGeneration.from_pretrained(model_name)
tokenizer.src_lang = "en"
tokenizer.tgt_lang = "fr"
def preprocess_function(examples):
    inputs = examples["English Sentence"]
    targets = examples["French Sentence"]
    model_inputs = tokenizer(inputs, max_length=128, truncation=True)
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True)
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
train_dataset = train_test_split['train'].map(preprocess_function, batched=True)
val_dataset = test_val_split['train'].map(preprocess_function, batched=True)
test_dataset = test_val_split['test'].map(preprocess_function,batched=True)
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)
training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",
    learning_rate=3e-5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    weight_decay=0.01,
    num_train_epochs=10,
    logging_dir="./logs",
    logging_steps=1000,
    save_strategy="steps"
)
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer
)
trainer.train()
```
